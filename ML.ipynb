Feature Engineering
Good features = good predictions.
Feature engineering:

Extracts hidden information

Captures market behavior

Converts finance knowledge into numbers

This step directly improves accuracy.

```python
#Daily Price Range
df['high_low_range'] = df['high'] - df['low']   # Highlow range captures daily price uncertainty

#Measures intraday volatility

#Larger range  more uncertainty
```

```python
# OpenClose Range: Shows bullish or bearish daily movement
df['open_close_range'] = df['close'] - df['open']
# if close is greater then open then the day was bulish and if the close is smaller then the open then the day is bearish
```

```python
# 4.3: Liquidity Ratio: Measures how actively a crypto is traded and Low liquidity  sudden volatility
df['liquidity_ratio'] = df['volume'] / df['marketCap']
# the formula used for liquidity ratio: volume/market cap
# If liquidity is high, it means many people are buying or selling the crypto on that day.
# High liquidity  more stable market
#Low liquidity  unstable, jumpy market
```

```python
#Moving Averages: The average price of the last N days
#Reduces sharp jumps

#Reduces random fluctuations

#Reveals the trend
df['ma_7'] = df.groupby('crypto_name')['close'].rolling(7).mean().reset_index(level=0, drop=True)
df['ma_14'] = df.groupby('crypto_name')['close'].rolling(14).mean().reset_index(level=0, drop=True)
```

```python
# rolling volatility
#Converts price movement into risk metric
#High volatility  prices jump a lot
#Low volatility  prices move smoothly
#Take the last 2 returns and calculate their standard deviation
# Rolling volatility is calculated as the standard deviation of returns over a moving time window. It measures how unstable the market is during that period and serves as the target variable for volatility prediction
df['return'] = df.groupby('crypto_name')['close'].pct_change()

df['rolling_volatility'] = (
    df.groupby('crypto_name')['return']
      .rolling(7)
      .std()
      .reset_index(level=0, drop=True)
)
```

```python
X = df.drop(columns=['rolling_volatility', 'date', 'timestamp'])
y = df['rolling_volatility']
```

```python
X = pd.get_dummies(X, columns=['crypto_name'], drop_first=True)
```

```python
split = int(len(X) * 0.7)

X_train = X.iloc[:split]
X_test  = X.iloc[split:]

y_train = y.iloc[:split]
y_test  = y.iloc[split:]
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)
```

```python
model.fit(X_train, y_train)
```

```python
import joblib
joblib.dump(model, "crypto_volatility_model.pkl")
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE :", mae)
print("RMSE:", rmse)
print("R Score:", r2)
#MAE (Mean Absolute Error): Average absolute difference between actual and predicted volatility
# Lower MAE  better model

#RMSE (Root Mean Squared Error):Penalizes large prediction errors more heavily
#Lower RMSE  fewer big mistakes
#R Score:How much variance in volatility is explained by the model
# R  1  excellent

#R  0  poor

#R < 0  very bad
# The trained Random Forest model was evaluated using MAE, RMSE, and R score on unseen test data to assess prediction accuracy and generalization performance.
```
